# Business Objective

Develop a retail application credit scorecard for predicting the probability of serious delinquency within the next 24 months.

The scorecard will support:

- Approval / Reject decisions
- Risk segmentation
- Cut-off determination
- Potential future pricing and limit assignment

1.1 Model Objective
Approval / Reject

1.2 Target Definition
Good = 0
Bad = 1
90+ DPD
24 month performance window

1.3 Population Definition
Retail customers
Individuals only
