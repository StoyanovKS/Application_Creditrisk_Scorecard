```python
import pandas as pd

def variable_summary(df):

    summary = []

    for col in df.columns:

        summary.append({
            "variable": col,
            "dtype": df[col].dtype,
            "count": df[col].count(),
            "missing_pct": round(df[col].isna().mean() * 100, 2),
            "min": df[col].min(),
            "p01": df[col].quantile(0.01),
            "p05": df[col].quantile(0.05),
            "p25": df[col].quantile(0.25),
            "median": df[col].median(),
            "p75": df[col].quantile(0.75),
            "p95": df[col].quantile(0.95),
            "p99": df[col].quantile(0.99),
            "max": df[col].max(),
            "mean": df[col].mean(),
            "std": df[col].std()
        })

    return pd.DataFrame(summary)

summary_df = variable_summary(df)

summary_df
```

```python
def scorecard_summary(df):

    summary = []

    for col in df.columns:

        summary.append({
            "variable": col,
            "missing_pct": round(df[col].isna().mean() * 100, 2),
            "unique_values": df[col].nunique(),
            "mean": df[col].mean(),
            "median": df[col].median(),
            "p95": df[col].quantile(0.95),
            "p99": df[col].quantile(0.99),
            "max": df[col].max()
        })

    return pd.DataFrame(summary)



numeric_cols = [
    'RevolvingUtilizationOfUnsecuredLines',
    'age',
    'NumberOfTime30-59DaysPastDueNotWorse',
    'DebtRatio',
    'MonthlyIncome',
    'NumberOfOpenCreditLinesAndLoans',
    'NumberOfTimes90DaysLate',
    'NumberRealEstateLoansOrLines',
    'NumberOfTime60-89DaysPastDueNotWorse',
    'NumberOfDependents'
]
```

## dictionary

```python
dictionary = pd.DataFrame({
    "variable": [
        "SeriousDlqin2yrs",
        "RevolvingUtilizationOfUnsecuredLines",
        "age",
        "NumberOfTime30-59DaysPastDueNotWorse",
        "DebtRatio",
        "MonthlyIncome",
        "NumberOfOpenCreditLinesAndLoans",
        "NumberOfTimes90DaysLate",
        "NumberRealEstateLoansOrLines",
        "NumberOfTime60-89DaysPastDueNotWorse",
        "NumberOfDependents"
    ],

    "data_type": [
        "Target",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric",
        "Numeric"
    ],

    "business_meaning": [
        "Target variable - serious delinquency within next 2 years",
        "Utilization ratio of unsecured credit lines",
        "Applicant age",
        "Number of times 30-59 days past due",
        "Debt to income ratio",
        "Monthly income",
        "Number of open credit lines and loans",
        "Number of times 90+ days past due",
        "Number of real estate loans or lines",
        "Number of times 60-89 days past due",
        "Number of dependents"
    ]
})
```

```python
dictionary["missing_pct"] = (
    df[dictionary["variable"]]
    .isna()
    .mean()
    .values
    * 100
).round(2)

dictionary["unique_values"] = [
    df[col].nunique()
    for col in dictionary["variable"]
]
```
