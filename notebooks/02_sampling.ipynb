# portfolio bad rate

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../data/raw/cs-training.csv")

df.head()
```

```python
target_dist = (
    df["SeriousDlqin2yrs"]
    .value_counts(normalize=True)
    .rename("pct")
    .mul(100)
    .round(2)
)

target_dist
```

```python
bad_rate = (
    df["SeriousDlqin2yrs"]
    .mean()
)

print(f"Bad Rate: {bad_rate:.2%}")
```

Portfolio Bad Rate ≈ 6.7%

The dataset is moderately imbalanced.

# train test split

```python
from sklearn.model_selection import train_test_split
train_dev, test = train_test_split(
    df,
    test_size=0.2,
    stratify=df["SeriousDlqin2yrs"],
    random_state=42
)
```

```python
print(train_dev["SeriousDlqin2yrs"].mean())
print(test["SeriousDlqin2yrs"].mean())
```

# split of train and validation sets

```python
train, validation = train_test_split(
    train_dev,
    test_size=0.25,
    stratify=train_dev["SeriousDlqin2yrs"],
    random_state=42
)
```

```python
print(train.shape)
print(validation.shape)
print(test.shape)
```

```python
# bad rate verification

for name, dataset in {
    "Train": train,
    "Validation": validation,
    "Test": test
}.items():

    print(
        f"{name}: "
        f"{dataset['SeriousDlqin2yrs'].mean():.4%}"
    )
```

```python
# analysis on bad rate

bad_rate_summary = pd.DataFrame({
    "dataset": ["Train", "Validation", "Test"],
    "records": [
        len(train),
        len(validation),
        len(test)
    ],
    "bad_rate": [
        train["SeriousDlqin2yrs"].mean(),
        validation["SeriousDlqin2yrs"].mean(),
        test["SeriousDlqin2yrs"].mean()
    ]
})

bad_rate_summary
```

Sampling Strategy

Method:
Stratified random sampling

Stratification Variable:
SeriousDlqin2yrs in order to preserve portfolio bad rate across all samples.

```python
train.to_csv(
    "../data/processed/train.csv",
    index=False
)

validation.to_csv(
    "../data/processed/validation.csv",
    index=False
)

test.to_csv(
    "../data/processed/test.csv",
    index=False
)
```
