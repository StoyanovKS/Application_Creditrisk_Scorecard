```python
# binning function

def fit_binning(df, variable, target):

    x = df[variable]
    y = df[target]

    optb = OptimalBinning(
        name=variable,
        dtype="numerical"
    )

    optb.fit(x, y)

    return optb
```

```python
candidate_variables = [
    "RevolvingUtilizationOfUnsecuredLines",
    "age",
    "NumberOfTime30-59DaysPastDueNotWorse",
    "DebtRatio",
    "MonthlyIncome",
    "NumberOfOpenCreditLinesAndLoans",
    "NumberOfTimes90DaysLate",
    "NumberRealEstateLoansOrLines",
    "NumberOfTime60-89DaysPastDueNotWorse",
    "NumberOfDependents"
]
```

```python
binning_models = {}

for variable in candidate_variables:

    print(f"Processing: {variable}")

    model = fit_binning(
        train,
        variable,
        "SeriousDlqin2yrs"
    )

    binning_models[variable] = model
```

```python
import pandas as pd
from optbinning import OptimalBinning

train = pd.read_csv("../data/processed/train.csv")

target = "SeriousDlqin2yrs"

output_file = "../data/outputs/binning_results.xlsx"
```

### Binning Conclusions

The following variables exhibit strong monotonic relationships and clear separation power:

- RevolvingUtilizationOfUnsecuredLines
- Age
- MonthlyIncome
- NumberOfTimes90DaysLate

Strong multicollinearity was confirmed among the delinquency variables:

- NumberOfTimes90DaysLate
- NumberOfTime30-59DaysPastDueNotWorse
- NumberOfTime60-89DaysPastDueNotWorse

Based on IV, monotonicity and business interpretability, NumberOfTimes90DaysLate was selected as the representative delinquency variable for further modelling.

The remaining two delinquency variables will be excluded from subsequent feature selection and logistic regression development.
