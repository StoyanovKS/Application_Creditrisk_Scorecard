```python
import shutil
import os

project_data = "./data/raw"

os.makedirs(project_data, exist_ok=True)

for file in os.listdir(path):
    shutil.copy(
        os.path.join(path, file),
        os.path.join(project_data, file)
    )

print("Files copied.")
```
